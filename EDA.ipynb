```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
os.chdir("./data")
train = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")
customeraccount = pd.read_csv("./customeraccount.csv")
transaction_header = pd.read_csv("./transaction_header.csv")
genel_kategoriler = pd.read_csv("./genel_kategoriler.csv")
customer = pd.read_csv("./customer.csv")
product_groups = pd.read_csv("./product_groups.csv")
transaction_sale = pd.read_csv("./transaction_sale/transaction_sale.csv")
```

General information of every csv file


```python
dfs = [train, test, customeraccount, transaction_header, genel_kategoriler,
       customer, product_groups, transaction_sale]
```

```python
def get_df_name(df):
    name = [x for x in globals() if globals()[x] is df][0]
    return name
```

```python
def dataframe_information(dataframes, head=5):
    for dataframe in dataframes:
        name_of_df = "=== "+get_df_name(dataframe)+" ==="
        print(name_of_df)
        print("Info")
        print(dataframe.info())
        print("Head")
        display(dataframe.head(head))
        print(dataframe.isnull().sum())
        print("Number of missing values: ", dataframe.isna().sum().sum())
        print("-"*20)
        print("\n\n\n")
```

```python
dataframe_information(dfs)
```

```python
for df in dfs:
    name_of_df = "=== "+get_df_name(df)+" ==="
    print(name_of_df)
    display(df.describe().T\
            .style.bar(subset=['mean'])\
            .background_gradient(subset=['std'], cmap='Greens')\
            .background_gradient(subset=['50%'], cmap='BuGn'))
    

```

```python
display(transaction_header)
```

```python

for df in dfs:
    name_of_df = "=== "+get_df_name(df)+" ==="
    print(name_of_df)
    corr = df.corr().round(2)
    plt.figure(figsize=(20,10))
    sns.heatmap(corr, vmin=-1, vmax=1, center=0, square=False, annot=True, cmap='coolwarm')
    plt.show()

```

```python
customeraccount = customeraccount.merge(customer, on="individualnumber")
transaction_header = transaction_header.merge(customeraccount, on="cardnumber")
transaction_sale = transaction_sale.merge(transaction_header, on="basketid")
train = train.merge(transaction_sale, on="individualnumber")
train
```

```python
corr = train.corr().round(2)
plt.figure(figsize=(20,10))
sns.heatmap(corr, vmin=-1, vmax=1, center=0, square=False, annot=True, cmap='coolwarm')
plt.show()
```
